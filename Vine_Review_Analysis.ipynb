```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
vine_df = pd.read_csv('vine_table.csv')
vine_df.head(10)
```

```python
vine_df.info()
```

```python
# 1. Filter the data and create a new DataFrame to retrieve all the rows where the total_votes count is equal to or greater than 20 to pick reviews that are more likely to be helpful and to avoid having division by zero errors later on.filtered_vine_df = vine_df[vine_df['total_votes'] >= 20]
filtered_vine_df = vine_df[vine_df['total_votes'] >= 20]
filtered_vine_df.head(10)
```

```python
# 2. Filter the new DataFrame or table created in Step 1 and create a new DataFrame or table to retrieve all the rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%.
new_vine_df = filtered_vine_df.loc[filtered_vine_df.helpful_votes / filtered_vine_df.total_votes >= 0.5]
new_vine_df.head(10)
```

```python
# 3. Filter the DataFrame or table created in Step 2, and create a new DataFrame or table that retrieves all the rows where a review was written as part of the Vine program (paid), vine == 'Y'.
vine_reviews = new_vine_df[new_vine_df['vine'] == 'Y']
vine_reviews.head(10)
```

```python
# 4. Repeat Step 3, but this time retrieve all the rows where the review was not part of the Vine program (unpaid), vine == 'N'.
not_vine_reviews = new_vine_df[new_vine_df['vine'] == 'N']
not_vine_reviews.head(10)
```

```python
# 5. Determine the total number of reviews that are part of vine program (paid)
count_vine = len(vine_reviews)
count_vine
```

```python
# 5. Determine the total number of reviews that are not part of the vine program (unpaid)
count_nonvine = len(not_vine_reviews)
count_nonvine
```

```python
# 5. Determine total number of 5-star reviews that are part of the vine program (paid)
max_rating_vine = len(vine_reviews[vine_reviews['star_rating']==5])
max_rating_vine
```

```python
# 5. Determine total number of 5-star reviews that are not part of the vine program (unpaid)
max_rating_nonvine = len(not_vine_reviews[not_vine_reviews['star_rating']==5])
max_rating_nonvine
```

```python
# 5. Determine percent of 5-star reviews that are part of the vine program (paid)
round((max_rating_vine/count_vine)*100, 1)
```

```python
# 5. Determine percent of 5-star reviews that are not part of the vine program (unpaid)
round((max_rating_nonvine/count_nonvine)*100,1)
```

```python
# EXTRA:  Determine average star rating for reviews that are part of the vine program (paid)
round(vine_reviews['star_rating'].mean(),1)
```

```python
# EXTRA:  Determine average star rating for reviews that are not part of the vine program (unpaid)
round(not_vine_reviews['star_rating'].mean(),1)
```
